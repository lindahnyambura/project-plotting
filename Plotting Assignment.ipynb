**A Histogram for CO2 Emissions**

```python
# A histogram that shows the distribution of carbon dioxide

plt.figure(figsize=(10, 6))
binwidth = 8
sns.histplot(data['co2'], bins=range(int(data['co2'].min()), int(data['co2'].max()) + binwidth, binwidth), kde=True)
plt.title('Distribution of CO2 Emissions')
plt.xlabel('CO2 Emissions')
plt.ylabel('Frequency')
plt.show()
```

Most vehicles have CO2 emissions around 300 grams per mile.
The distribution is right-skewed, with a long tail extending towards higher CO2 emissions. This suggests that while most vehicles have moderate emissions, there are a few with very high emissions.
The wide range of emissions indicates significant variability in the environmental impact of different vehicles.

**A Histogram for Engine Displacement**

```python
# A histogram for engine displacement(litres)

data = pd.read_csv(r'C:\Users\user\Project-Plotting\fuel_econ.csv')
plt.figure(figsize=(10, 6))
sns.histplot(data['displ'], bins=30, kde=True)
plt.title('Distribution of Engine Displacement (Liters)')
plt.xlabel('Engine Displacement (Liters)')
plt.ylabel('Frequency')
plt.show()
```

The histogram shows the distribution of engine displacement in litres for the vehicles in the dataset.
The peak indicates the most common engine displacement values. For example, if the peak is around 2 liters, it means that many vehicles have an engine displacement around this value.
The histogram is left-skewed, it suggests that most vehicles have smaller engine displacements, with fewer vehicles having larger engines.

**Histogram for City Fuel Economy**

```python
# histogram for city fuel economy

data = pd.read_csv(r'C:\Users\user\Project-Plotting\fuel_econ.csv')

plt.figure(figsize=(10, 6))
sns.histplot(data['city'], bins=30, kde=True)
plt.title('Distribution of City Fuel Economy (MPG)')
plt.xlabel('City Fuel Economy (MPG)')
plt.ylabel('Frequency')
plt.show()
```

The histogram shows the distribution of city fuel economy for the vehicles. 
The peak of the histogram indicates the most common city MPG values. For example, if the peak is around 20 MPG, it means that many vehicles have a city fuel economy around this value.
The histogram is right-skewed, it suggests that most vehicles have lower city MPG, with fewer vehicles achieving higher city MPG.

```python

from matplotlib.colors import LinearSegmentedColormap

data = pd.read_csv(r'C:\Users\user\Project-Plotting\fuel_econ.csv')

numeric_data = data.select_dtypes(include=[np.number])

colors = ["#C3553A", "#ffffff", "#3F7F93"]
cmap = LinearSegmentedColormap.from_list("custom_cmap", colors)

plt.figure(figsize=(12, 8))
correlation_matrix = numeric_data.corr()
sns.heatmap(correlation_matrix, annot=True, cmap=cmap, vmin=-1, vmax=1)
plt.title('Correlation Heatmap')
plt.show()
```

The heatmap visualizes the correlation between different numeric variables in the dataset. The values in each cell represent the correlation coefficient between the variables. These values range from -1, indicating a perfect negative correlation, to 1, indicating a perfect positive correlation.

Example of strong positive correlations
'city' (city fuel economy) and 'highway' (highway fuel economy) have a strong positive correlation, it suggests that vehicles with good city fuel economy also tend to have good highway fuel economy.

Example of strong negative correlations
'displ' (engine displacement) and 'ghgScore' (greenhouse gas score) have a strong negative correlation, it suggests that vehicles with larger engines tend to have lower GHG scores.
